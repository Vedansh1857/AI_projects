```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

##### Thus, we can see that this data is very unbalanced

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

# Comparing the values of both the transactions

```python
credit_card_data.groupby('Class').mean()
```

##### Here's a point to be noted that the mean of the amounts of the legit transactions is too much than that of fraudelent transactions

# Performing under-sampling for this unbalanced dataset

```python
# Building a sample dataset containing a similar distribution of the normal & fraudulent transactions
legit_sample = legit.sample(n = 492)
```

# Concatenating the two dataframes

```python
concatenated_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
concatenated_df.head()
```

```python
concatenated_df.tail()
```

```python
concatenated_df['Class'].value_counts()
```

```python
concatenated_df.groupby('Class').mean()
```

# Separating the features and labels of the dataset

```python
x = concatenated_df.drop(columns='Class', axis=1)
y = concatenated_df['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Creating & training the model

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

# Evaluating the model...

```python
# Firstly, calculating the predictions by our model on the training data
preds = model.predict(x_train)
```

```python
# checking the accuracy of our model on training set
acc_score = accuracy_score(y_train, preds)
print(acc_score)
```

```python
# Now, calculating the predictions by our model on the testing data
predictions = model.predict(x_test)
```

```python
# checking the accuracy of our model on testing set
acc = accuracy_score(y_test, predictions)
print(acc)
```

```python
print(classification_report(y_test, predictions))
```
